# Capstone Project - Predict House Sale Prices

## Introduction
### This data set is obtain from Kaggle Competition, and the original data source is from Ames Housing Dataset. The challenge of this project is the dataset contains many (81) variables. Hence, extensive data cleaning is needed before starting the data analysis process. The goal of this project is to predict the house prices as accurate as possible.
### This project utilizes three (3) features selection methods and six (6) Machine Larning Algorithms to predict the Ames House Price. All of three most common feature elimination methods have been explored, namely,  filter method, wrapper method, and embedded method. Two main branches of data analysis approach have been chosen, that is, the regularized method, and tree-based method. For the regularized method approach, we have applied three algorithms - Ridge regression, LASSO regression, and Elastic Net. For the tree-based approach, we have used three algorithms - Decision Tree, Random Forest, and XGBoost. The intention of these selections is to see how differnt feature selection methods and diffrenet machine learning algorithm influence the prediction power.
### We will utilize the RMSE (root-mean-squared-error) to evaluate and compare the prediction accuracy among the modeling results. Residual plot has also been used to see how the error varies within the range of predicted house price for each modeling result.
